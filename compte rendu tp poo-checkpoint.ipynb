```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Define a function to plot the correlation heatmap for a DataFrame
def plot_correlation_map(df):
    # Select numeric columns from the DataFrame
    numeric_columns = df.select_dtypes(include=['number'])
    
    # Calculate the correlation matrix
    corr = numeric_columns.corr()
    
    # Set up the size of the plot
    plt.figure(figsize=(12, 10))
    
    # Define the color palette for the heatmap
    cmap = sns.diverging_palette(220, 10, as_cmap=True)
    
    # Create the heatmap with annotations
    sns.heatmap(corr, cmap=cmap, square=True, cbar_kws={'shrink': 0.9}, annot=True, annot_kws={'fontsize': 12})
    
    # Display the plot
    plt.show()

# Call the function with the DataFrame 'df' to visualize the correlation heatmap
plot_correlation_map(df)
```

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Define the Title_Dictionary with title mappings
Title_Dictionary = {
    "Capt": "Officer",
    "Col": "Officer",
    "Major": "Officer",
    "Dr": "Officer",
    "Rev": "Officer",
    "Jonkheer": "Royalty",
    "Don": "Royalty",
    "Sir": "Royalty",
    "Lady": "Royalty",
    "the Countess": "Royalty",
    "Dona": "Royalty",
    "Mme": "Miss",
    "Mlle": "Miss",
    "Miss": "Miss",
    "Ms": "Mrs",
    "Mr": "Mr",
    "Mrs": "Mrs",
    "Master": "Master"
}

# Read the Titanic dataset into a DataFrame 'df' using ';' as the delimiter
df = pd.read_csv("titanic-passengers.csv", delimiter=';')

# Add a new 'Title' column based on the modified Title_Dictionary
df['Title'] = df['Name'].apply(lambda name: Title_Dictionary.get(name.split(',')[1].split('.')[0].strip(), "Unknown"))

# Use label encoding to convert 'Title' and 'Sex' into numerical representations
label_encoder = LabelEncoder()
df['Title'] = label_encoder.fit_transform(df['Title'])
df['Sex'] = label_encoder.fit_transform(df['Sex'])

# Calculate the correlation matrix
correlation_matrix = df[['Title', 'Sex', 'Fare', 'Age']].corr()

# Create a heatmap to visualize the correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
